```python
import openpyxl
import time

today_time=time.strftime("%Y%m%d")

#新規ブック作成
workbook=openpyxl.Workbook()
#Activesheet取得
worksheet=workbook.active

#請求書作成ここから
worksheet.cell(row=2,column=1).value="請求書"

#請求宛先
company_datas=[
    '株式会社ABC',
    '〒101-0022 東京都千代田区神田練塀町300',
    'TEL:03-1234-5678 FAX:03-1234-5678',
    '担当者名:鈴木一郎 様'
]

#請求書宛先転記
i=4
for company_data in company_datas:
    worksheet.cell(row=i,column=1).value=company_data
    i=i+1

#No表示
worksheet.cell(row=4,column=5).value="No."
worksheet.cell(row=4,column=6).value="0001"

#日付表示
worksheet.cell(row=5,column=5).value="日付"
worksheet.cell(row=5,column=6).value=time.strftime("%Y/%m/%d")


#商品データ
shopping_datas=[
    ['商品名','数量','単価','金額'],
    ['商品A',2,10000,20000],
    ['商品B',1,15000,15000],
]

#商品データ転記
for shopping_data in shopping_datas:
    worksheet.append(shopping_data)

Last_row=worksheet.max_row#最終行

#金額合計
sum=worksheet.cell(row=Last_row-1,column=4).value + worksheet.cell(row=Last_row,column=4).value
worksheet.cell(row=Last_row+1,column=4).value=sum

#行挿入
worksheet.insert_rows(8,2)
#最終行確認
Last_row=worksheet.max_row

#合計
worksheet.cell(row=Last_row+2,column=1).value="合計"
worksheet.cell(row=Last_row+2,column=4).value=sum

#消費税
tax=sum*0.10
worksheet.cell(row=Last_row+3,column=1).value="消費税"
worksheet.cell(row=Last_row+3,column=4).value=tax

#税込合計
Total=sum+tax
worksheet.cell(row=Last_row+4,column=1).value="税込合計"
worksheet.cell(row=Last_row+4,column=4).value=Total

#列挿入
worksheet.insert_cols(1)

#保存
workbook.save(f'請求書_{today_time}.xlsx')
```
